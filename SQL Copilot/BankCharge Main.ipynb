```python
cash_book = cash_book.drop(['DOCNO', 'DOCDATE', 'POSTDATE', 'TAXDATE', 'DESCRIPTION', 'DESCRIPTION2', 
                            'AREA', 'AGENT', 'PROJECT','JOURNAL', 'CHEQUENUMBER', 'CURRENCYCODE', 'CURRENCYRATE', 'DOCAMT', 'FROMDOCTYPE', 
                            'BOUNCEDDATE', 'GLTRANSID', 'CANCELLED', 'DEPOSITKEY', 'UPDATECOUNT', 'PRINTCOUNT', 
                            'ATTACHMENTS', 'NOTE'], axis=1)
cash_book_detail = cash_book_detail.drop(['DTLKEY', 'SEQ', 'AREA', 'AGENT', 'PROJECT', 
                                          'DESCRIPTION', 'GST_DOCDATE', 'GST_DOCNO', 'COMPANYNAME', 'REGISTERNO',
                                          'GSTNO', 'PERMITNO', 'COUNTRY', 'TAX', 'TARIFF', 'TAXRATE', 'TAXAMT',
                                          'LOCALTAXAMT', 'TAXINCLUSIVE', 'AMOUNT', 'LOCALAMOUNT', 'CURRENCYCODE',
                                          'CURRENCYAMOUNT'], axis=1)
merged_cash_book = pd.merge(cash_book, cash_book_detail, on='DOCKEY')
merged_cash_book
```

```python
# Drop rows with missing values in the target variable
merged_cash_book = merged_cash_book.dropna(subset=['BANKCHARGE'])
merged_cash_book = merged_cash_book.dropna(subset=['BANKCHARGEACCOUNT'])
merged_cash_book = merged_cash_book[merged_cash_book['DOCTYPE'] != 'OR']
merged_cash_book = merged_cash_book[merged_cash_book['CURRENCYRATE'] != 1.0]
merged_cash_book = merged_cash_book[merged_cash_book['CURRENCYRATE'] != 0.0]
merged_cash_book
```

```python
# Use OneHotEncoder instead of LabelEncoder for categorical features
pm_encoder = OneHotEncoder(sparse=False, drop='first')
code_encoder = OneHotEncoder(sparse=False, drop='first')

# Transform 'PAYMENTMETHOD' using one-hot encoding
payment_method_encoded = pm_encoder.fit_transform(merged_cash_book[['PAYMENTMETHOD']])
payment_method_columns = pm_encoder.get_feature_names_out(['PAYMENTMETHOD'])

# Transform 'CODE' using one-hot encoding
code_encoded = code_encoder.fit_transform(merged_cash_book[['CODE']])
code_columns = code_encoder.get_feature_names_out(['CODE'])

# Combine the one-hot encoded features with the other numerical feature
merged_cash_book_encoded = pd.concat([
    merged_cash_book[['LOCALDOCAMT', 'BANKCHARGE', 'CURRENCYRATE']].reset_index(drop=True),  # Reset index before concatenation
    pd.DataFrame(payment_method_encoded, columns=payment_method_columns).reset_index(drop=True),
    pd.DataFrame(code_encoded, columns=code_columns).reset_index(drop=True)
], axis=1)

# Display the DataFrame with the reset index
merged_cash_book_encoded
```

```python
# Split the data into features and target
features = merged_cash_book_encoded.drop('BANKCHARGE', axis=1)
target = merged_cash_book_encoded['BANKCHARGE']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Initialize and train the RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)
```

```python
# Evaluate the model
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predictions)
```

```python
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2): {r2}')
```

```python
# Now you can use the trained model to make predictions for new data
new_data = pd.DataFrame({
    'PAYMENTMETHOD': ['301-007'],
    'LOCALDOCAMT': [10358.12],
    'CODE': ['400-Y0002'],
    'CURRENCYRATE': [4.332000]
})

# Transform 'PAYMENTMETHOD', 'CODE', and 'CURRENCYRATE' using the previously fitted encoders
new_data_payment_method_encoded = pd.DataFrame(pm_encoder.transform(new_data[['PAYMENTMETHOD']]), columns=payment_method_columns)
new_data_code_encoded = pd.DataFrame(code_encoder.transform(new_data[['CODE']]), columns=code_columns)

# Combine the one-hot encoded features with the other numerical feature
new_data_encoded = pd.concat([
    new_data[['LOCALDOCAMT', 'CURRENCYRATE']],
    new_data_payment_method_encoded,
    new_data_code_encoded
], axis=1)

# Ensure new_data_encoded has all the necessary columns with zero values
missing_columns = set(X_train.columns) - set(new_data_encoded.columns)
for col in missing_columns:
    new_data_encoded[col] = 0

# Reorder columns to match the order during training
new_data_encoded = new_data_encoded[X_train.columns]

# Make predictions for the new data
prediction_new_data = model.predict(new_data_encoded)

print(f'Predicted BANKCHARGE for new data: {prediction_new_data[0]}')
```
