```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
cash_book_path = 'C:\\Users\\SQL\\Documents\\BOKOPT_GL_CB_202311291624.csv'
cash_book = pd.read_csv(cash_book_path)
cash_book
```

```python
cash_book_column = cash_book.columns
print(cash_book_column)
```

```python
cash_book = cash_book.drop(['DOCNO', 'DOCDATE', 'POSTDATE', 'TAXDATE', 'DESCRIPTION', 'DESCRIPTION2', 'AREA', 'AGENT', 'PROJECT', 'CURRENCYCODE', 'CURRENCYRATE', 'DOCAMT', 'FROMDOCTYPE', 'BOUNCEDDATE', 'GLTRANSID', 'CANCELLED', 'DEPOSITKEY', 'UPDATECOUNT', 'PRINTCOUNT', 'ATTACHMENTS', 'NOTE'], axis=1)
cash_book
```

```python
# Drop rows with missing values in the target variable
cash_book = cash_book.dropna(subset=['BANKCHARGE'])
cash_book = cash_book.dropna(subset=['BANKCHARGEACCOUNT'])
cash_book = cash_book[cash_book['DOCTYPE'] != 'PV']
cash_book
```

```python
# Use OneHotEncoder instead of LabelEncoder for categorical features
onehot_encoder = OneHotEncoder(sparse=False, drop='first')
```

```python
# Transform 'PAYMENTMETHOD' using one-hot encoding
payment_method_encoded = onehot_encoder.fit_transform(cash_book[['PAYMENTMETHOD']])
payment_method_columns = onehot_encoder.get_feature_names_out(['PAYMENTMETHOD'])
```

```python
# Combine the one-hot encoded features with the other numerical feature
features = pd.concat([cash_book[['LOCALDOCAMT']], pd.DataFrame(payment_method_encoded, columns=payment_method_columns)], axis=1)
target = cash_book['BANKCHARGE']
```

```python
# Drop any remaining rows with missing values from both features and target
features_target_combined = pd.concat([features, target], axis=1).dropna()
features = features_target_combined.iloc[:, :-1]
target = features_target_combined.iloc[:, -1]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=10)
X_train
```

```python
# Initialize and train the RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predictions)
```

```python
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2): {r2}')
```

```python
# Now you can use the trained model to make predictions for new data
new_data = pd.DataFrame({'PAYMENTMETHOD': ['301-001'],
                         'LOCALDOCAMT': [5350]})
new_data_encoded = pd.DataFrame(onehot_encoder.transform(new_data[['PAYMENTMETHOD']]), columns=payment_method_columns)
new_data = pd.concat([new_data[['LOCALDOCAMT']], new_data_encoded], axis=1)

# Ensure new_data has all the necessary columns with zero values
missing_columns = set(X_train.columns) - set(new_data.columns)
for col in missing_columns:
    new_data[col] = 0

# Reorder columns to match the order during training
new_data = new_data[X_train.columns]

prediction_new_data = model.predict(new_data)

print(f'Predicted BANKCHARGE for new data: {prediction_new_data[0]}')
```
