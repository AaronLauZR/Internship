```python
import numpy as np
import matplotlib.pyplot as plt

data = [1029.1, 4484.55, 9460.15, 15464.61, 17726.66, 21334.67, 22017.82, 24735.06, 25760.03, 26990.75, 26990.75, 26990.75, 31618.86, 32668.65, 33420.77, 38171.81, 41384.98, 42239.88, 42299.19, 48020.56, 52689.48, 54677.55, 55205.33, 56427.58, 58848.82, 59880.88, 64407.3, 64846.65, 65123.64, 67902.76, 68265.24, 68935.03, 72288.55, 78176.99, 82289.53, 87343.38, 95023.26, 97128.97, 100554.23, 131948.61, 142840.21, 153399.69]

mean_value = np.mean(data)
std_deviation = np.std(data)
rsd = (std_deviation / mean_value) * 100

# Calculate a range based on the relative standard deviation
lower_bound = mean_value - 2 * std_deviation
upper_bound = mean_value + 2 * std_deviation

if lower_bound <= 0:
    lower_bound = 0.01
elif upper_bound <= 0:
    upper_bound = 0.01

# Print the calculated values
print(f"Mean: {mean_value}")
print(f"Standard Deviation: {std_deviation}")
print(f"Relative Standard Deviation (RSD): {rsd:.2f}%")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Identify valid data within the bounds and outliers
valid_data = [x for x in data if lower_bound <= x <= upper_bound]
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Print valid data and outliers
print(f"\nValid Data: {valid_data}")
print(f"\nOutliers: {outliers}")

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data, label='Data')
plt.axhline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axhline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Lower Bound')
plt.axhline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper Bound')

# Highlight valid data and outliers
plt.scatter([data.index(x) for x in valid_data], valid_data, c='blue', label='Valid Data')
plt.scatter([data.index(x) for x in outliers], outliers, c='red', label='Outliers')

plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.title('Data with Mean, Valid Data, and Outliers')
plt.legend()
plt.show()
```
