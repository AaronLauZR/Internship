```python
import pandas as pd
```

```python
desc = 'C:/Users/SQL/Documents/categorized_expenses_desc.csv'
desc = pd.read_csv(desc)
desc
```

```python
# Specify the list of columns where you want to fill NaN with empty string
columns_to_fill = ['Who', 'What', 'Why', 'Where', 'When']

# Fill NaN with empty string in the specified columns
desc[columns_to_fill] = desc[columns_to_fill].fillna('')

desc = desc.drop(columns=['Unnamed: 0'])
desc
```

```python
user_selection = input(f"Enter Selection: ")

if user_selection in desc.columns:
    user_description = input(f"Enter Description: ")
    
    matching_rows = desc[desc[user_selection] == user_description]
    for index, row in matching_rows.iterrows():
            for column, value in row.items():
                print(f"{column}: {value}")
            print("\n-------------------\n")
    
    if matching_rows.empty:
        print(f"No rows found in '{user_selection}' with description '{user_description}'.")

else:
    print(f"No, the column '{user_selection}' does not exist in the DataFrame.")
```
