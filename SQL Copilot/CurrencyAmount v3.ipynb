```python
# Group by 'CODE' and collect 'CURRENCYAMOUNT' into lists
grouped_data = merged_cash_book.groupby(['CODE'])['CURRENCYAMOUNT'].apply(list).reset_index()

# Sort the 'CURRENCYAMOUNT' arrays in ascending order
grouped_data['CURRENCYAMOUNT'] = grouped_data['CURRENCYAMOUNT'].apply(lambda x: sorted(x))

# Display the resulting DataFrame
grouped_data
```

```python
# Function to plot interquartile range
def plot_iqr(code, data, median, q1, q3, outliers_low, outliers_high, valid_data):
    plt.figure(figsize=(10, 6))

    # Plot the data points
    plt.scatter(data, np.zeros_like(data), c='blue', label='All Data', alpha=0.7)

    # Highlight outliers
    plt.scatter(outliers_low, np.zeros_like(outliers_low), c='red', marker='o', label='Outliers Low', alpha=0.7)
    plt.scatter(outliers_high, np.zeros_like(outliers_high), c='red', marker='o', label='Outliers High', alpha=0.7)

    # Plot the median, Q1, Q3 as lines
    plt.plot([median, median], [0, 1], color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.plot([q1, q1], [0, 1], color='orange', linestyle='dashed', linewidth=2, label='Q1')
    plt.plot([q3, q3], [0, 1], color='purple', linestyle='dashed', linewidth=2, label='Q3')

    # Plot the valid data range
    plt.plot([min(valid_data), max(valid_data)], [0.5, 0.5], color='blue', linewidth=4, label='Valid Data Range')

    plt.title(f'Interquartile Range for CODE: {code}')
    plt.xlabel('CURRENCYAMOUNT')
    plt.yticks([])
    plt.legend()
    plt.show()
```

```python
# Function to calculate interquartile range
def calculate_iqr(data):
    # Convert 'decimal.Decimal' to float
    data = [float(val) for val in data]

    # Calculate the median, Q1, and Q3
    median = np.median(data)
    #median = np.mean(data)
    q1 = np.percentile(data, 5)
    q3 = np.percentile(data, 95)
    #q1 = median / 100 * 5
    #q3 = median / 100 * 195
    
    # Identify outliers
    outliers_low = [val for val in data if val < q1]
    outliers_high = [val for val in data if val > q3]

    # Filter valid data
    valid_data = [val for val in data if q1 <= val <= q3]

    return {
        "Median": median,
        "Q1": q1,
        "Q3": q3,
        "Data": data,
        "Outliers Low": outliers_low,
        "Outliers High": outliers_high,
        "Valid Data": valid_data
    }
```

```python
# User input
code_input = 'GST-KASTAM'
#code_input = '2000/C01'
#code_input = '9100/B18'
selected_data = grouped_data[grouped_data['CODE'] == code_input]['CURRENCYAMOUNT'].iloc[0]
iqr_result = calculate_iqr(selected_data)

# Display graph
plot_iqr(code_input, iqr_result['Data'], iqr_result['Median'], iqr_result['Q1'], iqr_result['Q3'],
         iqr_result['Outliers Low'], iqr_result['Outliers High'], iqr_result['Valid Data'])

# Display the result
print(f"Median: {iqr_result['Median']}")
print(f"Q1: {iqr_result['Q1']}")
print(f"Q3: {iqr_result['Q3']}")
print(f"Data: {iqr_result['Data']}")
print(f"Outliers Low: {iqr_result['Outliers Low']}")
print(f"Outliers High: {iqr_result['Outliers High']}")
print(f"Valid Data: {iqr_result['Valid Data']}")
```
