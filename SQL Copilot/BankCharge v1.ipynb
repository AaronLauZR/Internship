```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
cash_book_path = 'C:\\Users\\SQL\\Documents\\BOKOPT_GL_CB_202311291624.csv'
cash_book = pd.read_csv(cash_book_path)
cash_book
```

```python
cash_book_column = cash_book.columns
print(cash_book_column)
```

```python
cash_book = cash_book.drop(['DOCNO', 'DOCDATE', 'DOCTYPE', 'POSTDATE', 'TAXDATE', 'DESCRIPTION', 'DESCRIPTION2', 'AREA', 'AGENT', 'PROJECT', 'CURRENCYCODE', 'CURRENCYRATE', 'DOCAMT', 'FROMDOCTYPE', 'BOUNCEDDATE', 'GLTRANSID', 'CANCELLED', 'DEPOSITKEY', 'UPDATECOUNT', 'PRINTCOUNT', 'ATTACHMENTS', 'NOTE'], axis=1)
cash_book
```

```python
# Drop rows with missing values in the target variable
cash_book = cash_book.dropna(subset=['BANKCHARGE'])
cash_book = cash_book.dropna(subset=['BANKCHARGEACCOUNT'])
cash_book
```

```python
label_encoder = LabelEncoder()
cash_book['PAYMENTMETHOD'] = label_encoder.fit_transform(cash_book['PAYMENTMETHOD'])
```

```python
# Select relevant features and target variable
features = cash_book[['PAYMENTMETHOD', 'LOCALDOCAMT']]
target = cash_book['BANKCHARGE']
```

```python
# Convert categorical features to numerical using one-hot encoding
features = pd.get_dummies(features, columns=['PAYMENTMETHOD'], drop_first=True)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=10)
```

```python
# Initialize and train the RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predictions)
```

```python
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2): {r2}')
```

```python
# Now you can use the trained model to make predictions for new data
new_data = pd.DataFrame({'PAYMENTMETHOD': [301-100],
                         'LOCALDOCAMT': [5350.0]})
new_data = pd.get_dummies(new_data, columns=['PAYMENTMETHOD'], drop_first=True)

# Ensure new_data has all the necessary columns with zero values
missing_columns = set(X_train.columns) - set(new_data.columns)
for col in missing_columns:
    new_data[col] = 0

# Reorder columns to match the order during training
new_data = new_data[X_train.columns]

prediction_new_data = model.predict(new_data)

print(f'Predicted BANKCHARGE for new data: {prediction_new_data[0]}')
```
